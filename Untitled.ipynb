Reasons of each 911 calls

```python
df["Reasons"] = df["title"].apply(lambda x:x.split(":")[0])
```

Counts and their plot of these calls

```python
sns.countplot(x="Reasons" ,data = df)
```

```python


#converting the strings date time to DayTime Objects
df['timeStamp'] = pd.to_datetime(df["timeStamp"])
```

```python



type(df["timeStamp"].iloc[0])
```

```python
#converting these timestamps to their respective month,day,hours,date
df["month"] = df["timeStamp"].apply(lambda time :time.month)
df["day"] = df["timeStamp"].apply(lambda time: time.dayofweek)
df["hours"] = df["timeStamp"].apply(lambda time: time.hour)
df["date"] = df['timeStamp'].apply(lambda time:time.date())
```

```python

okk = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df["week"] = df["day"].map(okk)
```

Daywise analysis of these calls

```python

sns.countplot(x='week',data=df,hue='Reasons',palette='magma')
plt.legend(loc = 3)
```



Monthwise Analysis of these calls

```python
sns.countplot(x='month',data=df,hue='Reasons',palette='winter')
plt.legend(loc = 3)
```

```python
ohh = df.groupby("month").count() #gives count of each instances of each column for that particular month
ohh.head(5)
```

Analysing calls getting for different months

```python
ohh["twp"].plot()
```

```python
sns.lmplot(x = "month" , y="twp" , data = ohh.reset_index())
```

```python
#creating data set for analysis calls with hours

dayHour = df.groupby(by=['week','hours']).count()['Reasons'].unstack()
dayHour.head()
```

Analysing the variation of calls with hours on different days using heatmaps and other maps like clustermaps.

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour,cmap='magma_r')
```

```python
sns.clustermap(dayHour,cmap='magma_r')
```

Analysing the variation of calls with months on different days using heatmaps and other maps like clustermaps.

```python


dayMonth = df.groupby(by=['week','month']).count()['Reasons'].unstack()
dayMonth.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayMonth,cmap='viridis')
```

```python
sns.clustermap(dayMonth,cmap='viridis')
```
